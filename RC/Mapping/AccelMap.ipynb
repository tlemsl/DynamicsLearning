```python
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
d = np.array(accel_map)
plt.plot(d.T[1], d.T[0],"-ro")
plt.ylabel('Accel input')
plt.xlabel('Actual velocity')
# plt.legend()

plt.show()
d.T
```

```python
from sklearn.linear_model import LinearRegression

accel_map = np.array(accel_map)

# Extract actual and pseudo steer values
actual_accel = accel_map[:, 1].reshape(-1, 1)  # Actual steer (reshape for sklearn)
pseudo_accel = accel_map[:, 0].reshape(-1, 1)  # Pseudo steer

# Linear regression: Actual -> Pseudo
model_actual_to_pseudo = LinearRegression()
model_actual_to_pseudo.fit(actual_accel, pseudo_accel)

# Linear regression: Pseudo -> Actual
model_pseudo_to_actual = LinearRegression()
model_pseudo_to_actual.fit(pseudo_accel, actual_accel)

# Get the model parameters
slope_actual_to_pseudo = model_actual_to_pseudo.coef_[0][0]
intercept_actual_to_pseudo = model_actual_to_pseudo.intercept_[0]

slope_pseudo_to_actual = model_pseudo_to_actual.coef_[0][0]
intercept_pseudo_to_actual = model_pseudo_to_actual.intercept_[0]

# Print the results
print(f"Mapping function (Actual to Pseudo): Pseudo = {slope_actual_to_pseudo:.4f} * Actual + {intercept_actual_to_pseudo:.4f}")
print(f"Mapping function (Pseudo to Actual): Actual = {slope_pseudo_to_actual:.4f} * Pseudo + {intercept_pseudo_to_actual:.4f}")

# Plotting Actual to Pseudo
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(actual_accel, pseudo_accel, 'o', label='Data Points')
plt.plot(actual_accel, model_actual_to_pseudo.predict(actual_accel), 'r-', label='Fitted Line')
plt.xlabel('Actual Aceel')
plt.ylabel('Pseudo Aceel')
plt.title('Actual Aceel to Pseudo Accel')
plt.legend()
plt.grid(True)

# Plotting Pseudo to Actual
plt.subplot(1, 2, 2)
plt.plot(pseudo_accel, actual_accel, 'o', label='Data Points')
plt.plot(pseudo_accel, model_pseudo_to_actual.predict(pseudo_accel), 'r-', label='Fitted Line')
plt.xlabel('Pseudo Aceel')
plt.ylabel('Actual Aceel')
plt.title('Pseudo Aceel to Actual Aceel')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

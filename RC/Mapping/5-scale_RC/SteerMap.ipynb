```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
steer_map = []
```

```python
# Load the dataset
file_path = 'steer_map_-500.csv'  # replace with the actual path to your dataset
data = pd.read_csv(file_path)

pseudo_steer = data["steer"]
v = data["v_horizontal"]
x = data["x"]
y = data["y"]
yaw = data["yaw"]
```

```python
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(pseudo_steer,"-ro",  label="steer")

plt.legend()

plt.show()
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(yaw,"-bo",  label="yaw")

plt.legend()

plt.show()
```

```python
steering_start = 170
steering_end = 205
filtered_pseudo_steer = pseudo_steer[steering_start:steering_end]
filtered_yaw = yaw[steering_start:steering_end]
filtered_v = v[steering_start:steering_end]
```

```python
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(filtered_pseudo_steer,"-ro",  label="steer")

plt.legend()

plt.show()
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(filtered_yaw,"-bo",  label="yaw")

plt.legend()

plt.show()
```

```python
# Define vehicle wheelbase (L), this value is typically known
L = 0.43  # Example wheelbase in meters, replace with actual value

# Calculate change in yaw (yaw rate) using np.gradient
dt = 0.1  # Assuming 1-second intervals, adjust if necessary
yaw_rate = np.gradient(filtered_yaw, dt)

# Calculate actual steering angle
actual_steer = np.arctan(yaw_rate * L / filtered_v)

steer_map.append([actual_steer.mean(), filtered_pseudo_steer.mean()])
steer_map
```

```python
from sklearn.linear_model import LinearRegression
# Convert to numpy array for easier manipulation
steer_map = np.array(steer_map)

# Extract actual and pseudo steer values
actual_steer = steer_map[:, 0].reshape(-1, 1)  # Actual steer (reshape for sklearn)
pseudo_steer = steer_map[:, 1].reshape(-1, 1)  # Pseudo steer

# Linear regression: Actual -> Pseudo
model_actual_to_pseudo = LinearRegression()
model_actual_to_pseudo.fit(actual_steer, pseudo_steer)

# Linear regression: Pseudo -> Actual
model_pseudo_to_actual = LinearRegression()
model_pseudo_to_actual.fit(pseudo_steer, actual_steer)

# Get the model parameters
slope_actual_to_pseudo = model_actual_to_pseudo.coef_[0][0]
intercept_actual_to_pseudo = model_actual_to_pseudo.intercept_[0]

slope_pseudo_to_actual = model_pseudo_to_actual.coef_[0][0]
intercept_pseudo_to_actual = model_pseudo_to_actual.intercept_[0]

# Print the results
print(f"Mapping function (Actual to Pseudo): Pseudo = {slope_actual_to_pseudo:.4f} * Actual + {intercept_actual_to_pseudo:.4f}")
print(f"Mapping function (Pseudo to Actual): Actual = {slope_pseudo_to_actual} * Pseudo + {intercept_pseudo_to_actual}")

# Plotting Actual to Pseudo
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(actual_steer, pseudo_steer, 'o', label='Data Points')
plt.plot(actual_steer, model_actual_to_pseudo.predict(actual_steer), 'r-', label='Fitted Line')
plt.xlabel('Actual Steering')
plt.ylabel('Pseudo Steering')
plt.title('Actual Steering to Pseudo Steering')
plt.legend()
plt.grid(True)

# Plotting Pseudo to Actual
plt.subplot(1, 2, 2)
plt.plot(pseudo_steer, actual_steer, 'o', label='Data Points')
plt.plot(pseudo_steer, model_pseudo_to_actual.predict(pseudo_steer), 'r-', label='Fitted Line')
plt.xlabel('Pseudo Steering')
plt.ylabel('Actual Steering')
plt.title('Pseudo Steering to Actual Steering')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

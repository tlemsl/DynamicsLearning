```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
steer_map = []
L = 0.26  # Example wheelbase in meters, replace with actual value
```

```python
# Load the dataset
file_path = 's_-400.csv'  # replace with the actual path to your dataset
data = pd.read_csv(file_path)

pwm_steer = data["steering_angle"]
v = data["v_horizontal"]
x = data["x"]
y = data["y"]
yaw = data["yaw"]
omega = data["omega"]
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(pwm_steer,"-ro",  label="steer")

plt.legend()

plt.show()
plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(yaw,"-bo",  label="yaw")

plt.legend()

plt.show()
```

```python
steering_start = 0
steering_end = -1
actual_pwm_steer =-400
filtered_v = v[steering_start:steering_end]
filtered_omega = omega[steering_start:steering_end]

plt.figure(figsize = (10, 5))
# plt.plot(accel,"-bo",  label="accel")
plt.plot(filtered_omega,"-bo",  label="omega")

plt.legend()

plt.show()
```

```python

# Calculate actual steering angle
actual_steer = np.arctan(omega * L / filtered_v)

steer_map.append([actual_steer.mean(), actual_pwm_steer])
steer_map
```

```python
from sklearn.linear_model import LinearRegression
# Convert to numpy array for easier manipulation
steer_map = np.array(steer_map)

# Extract actual and pseudo steer values
actual_steer = steer_map[:, 0].reshape(-1, 1)  # Actual steer (reshape for sklearn)
pwm_steer = steer_map[:, 1].reshape(-1, 1)  # Pseudo steer

# Linear regression: Actual -> PWM
model_actual_to_pwm = LinearRegression()
model_actual_to_pwm.fit(actual_steer, pwm_steer)

# Linear regression: PWM -> Actual
model_pwm_to_actual = LinearRegression()
model_pwm_to_actual.fit(pwm_steer, actual_steer)

# Get the model parameters
slope_actual_to_pwm = model_actual_to_pwm.coef_[0][0]
intercept_actual_to_pwm = model_actual_to_pwm.intercept_[0]

slope_pwm_to_actual = model_pwm_to_actual.coef_[0][0]
intercept_pwm_to_actual = model_pwm_to_actual.intercept_[0]

# Print the results
print(f"Mapping function (Actual to PWM): PWM = {slope_actual_to_pwm:.4f} * Actual + {intercept_actual_to_pwm:.4f}")
print(f"Mapping function (PWM to Actual): Actual = {slope_pwm_to_actual} * PWM + {intercept_pwm_to_actual}")

# Plotting Actual to Pseudo
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(actual_steer, pwm_steer, 'o', label='Data Points')
plt.plot(actual_steer, model_actual_to_pwm.predict(actual_steer), 'r-', label='Fitted Line')
plt.xlabel('Actual Steering')
plt.ylabel('PWM')
plt.title('Actual Steering to PWM')
plt.legend()
plt.grid(True)

# Plotting Pseudo to Actual
plt.subplot(1, 2, 2)
plt.plot(pwm_steer, actual_steer, 'o', label='Data Points')
plt.plot(pwm_steer, model_pwm_to_actual.predict(pwm_steer), 'r-', label='Fitted Line')
plt.xlabel('PWM')
plt.ylabel('Actual Steering')
plt.title('PWM to Actual Steering')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

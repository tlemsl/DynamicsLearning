```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PWM_to_RPM = {}
```

```python
PWM_to_RPM[50] = 894
PWM_to_RPM[100] = 1407
PWM_to_RPM[150] = 1698
PWM_to_RPM[200] = 1884
PWM_to_RPM[250] = 1955
```

```python
def plot_pwm_to_rpm():
    plt.plot(PWM_to_RPM.keys(), PWM_to_RPM.values())
    plt.xlabel('PWM')
    plt.ylabel('RPM')
    plt.title('PWM to RPM')
    plt.show()

plot_pwm_to_rpm()
```
